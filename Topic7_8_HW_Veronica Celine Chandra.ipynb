##### Nomor 3a:

```python
fig = plt.figure(figsize=(10, 8))

plt.subplot(211)
plt.title('Massa Tubuh vs Panjang Sayap')
ax1 = sns.scatterplot(data = df3, x = 'body_mass_g', y = 'flipper_length_mm', hue = 'island')
ax1.set_xlabel("Massa Tubuh (Gram)")
ax1.set_ylabel("Panjang Sayap (Milimeter)")

plt.subplot(212)
plt.title('Massa Tubuh vs Panjang Paruh')
ax2 = sns.scatterplot(data = df3, x = 'body_mass_g', y = 'bill_length_mm', hue = 'island')
ax2.set_xlabel("Massa Tubuh (Gram)")
ax2.set_ylabel("Panjang Paruh (Milimeter)")


plt.tight_layout()
plt.show()
```

##### Nomor 3b:
Kolom `body_mass_g` lebih baik digunakan untuk memprediksi `flipper_length_mm`.
